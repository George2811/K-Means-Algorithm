```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data =  pd.read_csv('1fXr31hcEemkYxLyQ1aU1g_50fc36ee697c4b158fe26ade3ec3bc24_Banknote-authentication-dataset-.csv')
```

```python
v1_mean = np.mean(data['V1'])
v2_mean = np.mean(data['V2'])

v1_std = np.std(data['V1'])
v2_std = np.std(data['V2'])

print("Variance")
print("Mean:", v1_mean)
print("Standard Deviation:", v1_std)
print("************")
print("Skewness")
print("Mean:", v2_mean)
print("Standard Deviation:", v2_std)
```

```python
variance = data['V1']
skewness = data['V2']

plt.xlabel('Variance')
plt.ylabel('Skewness')
plt.scatter(variance, skewness)
```

```python
from sklearn.cluster import KMeans
variance_skewness = np.column_stack((variance, skewness))

km_res = KMeans(n_clusters=3).fit(variance_skewness)

clusters = km_res.cluster_centers_

plt.xlabel('Variance')
plt.ylabel('Skewness')
plt.scatter(variance, skewness)
plt.scatter(clusters[:,0], clusters[:,1], s = 2000, alpha=0.45)
```
